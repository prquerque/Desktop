```python

from flask import Flask, jsonify, request
from flask_ngrok import run_with_ngrok

app = Flask(__name__)
run_with_ngrok(app)

# Banco de dados em mem√≥ria (lista de dicion√°rios)
usuarios = [
    {"id": 1, "nome": "Ana", "email": "ana@email.com"},
    {"id": 2, "nome": "Carlos", "email": "carlos@email.com"}
]

# üîπ GET - Listar todos os usu√°rios
@app.route("/usuarios", methods=["GET"])
def listar_usuarios():
    return jsonify(usuarios)

# üîπ GET - Buscar usu√°rio por ID
@app.route("/usuarios/<int:id>", methods=["GET"])
def buscar_usuario(id):
    for usuario in usuarios:
        if usuario["id"] == id:
            return jsonify(usuario)
    return jsonify({"erro": "Usu√°rio n√£o encontrado"}), 404

# üîπ POST - Criar novo usu√°rio
@app.route("/usuarios", methods=["POST"])
def criar_usuario():
    novo_usuario = request.get_json()
    novo_usuario["id"] = usuarios[-1]["id"] + 1 if usuarios else 1
    usuarios.append(novo_usuario)
    return jsonify(novo_usuario), 201

# üîπ PUT - Atualizar usu√°rio
@app.route("/usuarios/<int:id>", methods=["PUT"])
def atualizar_usuario(id):
    dados = request.get_json()
    for usuario in usuarios:
        if usuario["id"] == id:
            usuario["nome"] = dados.get("nome", usuario["nome"])
            usuario["email"] = dados.get("email", usuario["email"])
            return jsonify(usuario)
    return jsonify({"erro": "Usu√°rio n√£o encontrado"}), 404

# üîπ DELETE - Remover usu√°rio
@app.route("/usuarios/<int:id>", methods=["DELETE"])
def deletar_usuario(id):
    for usuario in usuarios:
        if usuario["id"] == id:
            usuarios.remove(usuario)
            return jsonify({"mensagem": "Usu√°rio removido com sucesso"})
    return jsonify({"erro": "Usu√°rio n√£o encontrado"}), 404

app.run()
```
